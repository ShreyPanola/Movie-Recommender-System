<h2>Main (Content-Based) Recommender System</h2>

```python
# Selecting and aggregating useful data columns.
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.head()
```

```python
# Checking for null values.
movies.isnull().sum()
```

```python
# As count for null values is less, we are dropping the null values.
movies.dropna(inplace=True)

movies.isnull().sum()
```

```python
import ast # For converting string format of dictionary(JSON format) to actual dictionary.
```

```python
# Function desc: Extracting list of values from dictionary where column name == 'name'.
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
# Extracting features from genres dictionary.
movies['genres']=movies['genres'].apply(convert)
movies.head()
```

```python
# Extracting features from keywords dictionary.
movies['keywords']=movies['keywords'].apply(convert)
movies.head()
```

```python
# Function desc: Creating list of top 3 cast from cast dictionary where value for name == 'name'.
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
```

```python
# Applying convert3 function.
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
# Function desc: Fetching director from crew dictionary where value for job == 'Director'.
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
# Applying fetch_director function.
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
# Final result of extracting important features from generes, keywords, cast and crew columns.
movies.head()
```

```python
# Starting preprocessing on overview column.
# Getting list of works from these sentence strings.
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#Function desc: To remove spaces between different words. For eg. James Camron -> JamesCamron
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
# Creating unique list for genres for further use in Genre(Top 10) recommender section.
genres=[]
for i in movies['genres']:
  for j in i:
    if j not in genres:
      genres.append(j)
print(genres)
print(len(genres))
```

```python
movies.sample(5)
```

```python
# Creating common column 'tags' from below columns.
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
# Dropping unnecessary columns and creating a new dataframe 'new_df'.
new_df = movies.drop(columns=['overview','genres','keywords','cast','crew'])
new_df.head()
```

```python
# Combining all features in one string.
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
new_df.head()
```

```python
# Applying lower case so that same words words(Eg. Action,action) are not considered separate.
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
#Stemming: For limiting the occurance of same meaning words over different variations. (For eg act,action,actions -> act)
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
# Creating custom stem function for stemming.
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
print("Before: ", new_df['tags'][0])
print("After: ",stem(new_df['tags'][0]))
```

```python
# Finally applying stem function on tags column.
new_df['tags']=new_df['tags'].apply(stem)
new_df['tags']
```

```python
# Importing count vectorizer from scikit learn library. Bag of words technique to be used.
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=40000, stop_words="english")#Here max features indicate maximum features(words) to consider. stop_words indicates to remove words like a,an,the,prepositions which help for sentence formation and doesn't add any meaning.
```

```python
# We will now apply CountVectorizer function to the tags column and find out most occuring words from these tags.
vectors = cv.fit_transform(new_df['tags']).toarray()
vectors
```

```python
for i in cv.get_feature_names_out():
  print(i)
```

```python
# Importing cosine_similarity from scikit learn library
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Finding similarities between each movie pair 5000x5000.
similarity = cosine_similarity(vectors)
similarity
```

```python
#recommend: Core function to display result on basis of movie title input
def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:6]
  for i in movies_list:
    print(new_df.iloc[i[0]]. title)
```

```python
recommend(new_df.iloc[0].title)
```
